```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
retail_df = pd.read_csv("Retail Transactions Dataset-4.csv", sep = "[\"*;*]", engine='python', header=None, skiprows=1)
retail_df.head(10)
```

```python
retail_df = retail_df[[1,4]]
retail_df.rename(columns={1:'invoice_no', 4:'product_no'},inplace=True)
retail_df.head(10)
```

```python
retail_df.shape
```

```python
retail_df.replace(to_replace=r'[^0-9]', value='', regex=True, inplace=True)
retail_df.head(20)
```

```python
retail_df.sort_values(by=['invoice_no','product_no'], inplace=True)
retail_df.head(20)
```

```python
retail_df['product_no'].value_counts().sort_index()
```

```python
retail_df.duplicated().sum()
```

```python
retail_df.drop_duplicates(inplace=True, ignore_index=True)
retail_df['product_no'].value_counts().sort_index()
```

```python
retail_df.head(20)
```

```python
allinvoices = []
newinvoice = []
last_invoice_no = ''
for i in range(len(retail_df)):
    if (retail_df.loc[i,'invoice_no'] != last_invoice_no):
        if (len(newinvoice) > 0):
            allinvoices.append(newinvoice)
        newinvoice = []
        newinvoice.append(retail_df.loc[i,'product_no'])
        last_invoice_no = retail_df.loc[i,'invoice_no']
    else:
        newinvoice.append(retail_df.loc[i,'product_no'])

print(allinvoices)
```

```python
te = TransactionEncoder()
te_ary = te.fit(allinvoices).transform(allinvoices)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
df_rules_by_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
df_rules_by_confidence.sort_values(by='confidence', ascending=False)
```

```python
df_rules_by_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
df_rules_by_lift.sort_values(by='lift', ascending=False)
```
